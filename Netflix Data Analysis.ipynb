```python
# Standardize column names
df.columns = [c.strip() for c in df.columns]
```

```python
# Rename observed columns to consistent names:
# 'Category' -> content_type (Movie/TV Show), 'Type' -> listed_in (genres), 'Release_Date' -> date_added
df.rename(columns={'Category':'content_type', 'Type':'listed_in', 'Release_Date':'date_added'}, inplace=True)
```

```python
# Parse date_added -> datetime and extract year_for_trend
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_for_trend'] = df['date_added'].dt.year
if 'Release_Year' in df.columns:
    df['year_for_trend'] = df['year_for_trend'].fillna(pd.to_numeric(df['Release_Year'], errors='coerce'))
df['year_for_trend'] = df['year_for_trend'].astype('Int64')
```

```python
# Parse Duration (minutes or seasons)
def parse_duration_val(x):
    if pd.isnull(x): return np.nan
    s = str(x)
    if 'min' in s:
        try: return int(s.split()[0])
        except: return np.nan
    if 'Season' in s:
        try: return int(s.split()[0])
        except: return np.nan
    return np.nan

df['duration_parsed'] = df['Duration'].apply(parse_duration_val)
```

```python
# Genres: split the 'listed_in' column into lists
df['listed_in'] = df['listed_in'].fillna('Unknown')
df['genres_list'] = df['listed_in'].apply(lambda x: [g.strip() for g in str(x).split(',') if g.strip()!=''])
```

```python
# Countries: split 'Country' into list
if 'Country' in df.columns:
    df['Country'] = df['Country'].fillna('Unknown')
    df['country_list'] = df['Country'].apply(lambda x: [c.strip() for c in str(x).split(',') if c.strip()!=''])
else:
    df['country_list'] = [[] for _ in range(len(df))]
```

```python
# Content type normalization
df['content_type'] = df['content_type'].fillna('Unknown')
```

```python
# Summary
summary = {
    'total_records': int(len(df)),
    'unique_titles': int(df['Title'].nunique() if 'Title' in df.columns else len(df)),
    'years_range': f"{int(df['year_for_trend'].min())} - {int(df['year_for_trend'].max())}" if df['year_for_trend'].notna().any() else "Unknown"
}
print("Summary:", summary)
```

```python
# Plot for Movies vs TV Shows over years
```

```python

df_time = df.dropna(subset=['year_for_trend']).copy()
type_year = df_time.groupby(['year_for_trend','content_type']).size().unstack(fill_value=0).sort_index()

plt.figure(figsize=(10,5))
for col in type_year.columns:
    plt.plot(type_year.index.astype(int), type_year[col], marker='o', label=col)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Movies vs TV Shows Over Years (by date_added year)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot for Top 10 genres overall 
genres_exploded = df[['Title','genres_list']].explode('genres_list')
genres_exploded['genres_list'] = genres_exploded['genres_list'].fillna('Unknown')
genre_counts = genres_exploded['genres_list'].value_counts().rename_axis('genre').reset_index(name='count')

top10 = genre_counts.head(10)
plt.figure(figsize=(10,6))
plt.bar(top10['genre'], top10['count'])
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Top 10 Genres in Netflix Catalog (Overall)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plot for Trends of top 5 genres 
top5_genres = genre_counts['genre'].head(5).tolist()
genre_year_ts = genres_exploded.merge(df[['Title','year_for_trend']], on='Title', how='left')
genre_year_ts = genre_year_ts.dropna(subset=['year_for_trend']).copy()
genre_year_ts['year_for_trend'] = genre_year_ts['year_for_trend'].astype(int)
genre_year_pivot = genre_year_ts[genre_year_ts['genres_list'].isin(top5_genres)].groupby(['year_for_trend','genres_list']).size().unstack(fill_value=0).sort_index()

plt.figure(figsize=(10,6))
for g in genre_year_pivot.columns:
    plt.plot(genre_year_pivot.index, genre_year_pivot[g], marker='o', label=g)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Trends of Top 5 Genres Across Years')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Plot for Top 15 countries by content count 
countries_exploded = df[['Title','country_list']].explode('country_list')
countries_exploded['country_list'] = countries_exploded['country_list'].fillna('Unknown')
country_counts = countries_exploded['country_list'].value_counts().rename_axis('country').reset_index(name='count')

top15 = country_counts.head(15)
plt.figure(figsize=(12,6))
plt.bar(top15['country'], top15['count'])
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 15 Countries by Content Count on Netflix')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
